```python
import sklearn
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Gráfica de la varianza explicada acumulada.
def cumulative_explained_variance_plot(expl_variance):
    cum_var_exp = np.cumsum(expl_variance)
    plt.figure(dpi = 100, figsize = (8, 6))
    plt.title('Curva acumulativa de la varianza explicada VS n° de componentes principales', 
            fontdict= dict(family ='serif', size = 16))
    plt.xlabel('Número de componentes principales',
             fontdict= dict(family ='serif', size = 14))
    plt.ylabel('Varianza explicada acumulativa',
             fontdict= dict(family ='serif', size = 14))  
    nc = np.arange(1, expl_variance.shape[0] + 1)
    plt.plot(nc, cum_var_exp, '--r')
    plt.plot(nc, cum_var_exp, 'b*', ms = 5)   
    plt.show()
```

```python
Datos = np.array([ [98,81,38], [103,84,38], [103,86,42], [105,86,42], [109,88,44], [123,95,46],
     [134,100,48], [136,102,49], [123,92,50], [133,99,51], [133,102,51] , [133,102,51],
     [138,98,51], [138,99,51], [141,105,53], [149,107,55], [153,107,56], [147,108,57], 
     [155,117,60], [158,115,62], [155,115,63], [159,118,63], [162,124,61], [177,132,67]])
```

# PCA con Datos sin estandarizar 

```python
from sklearn.decomposition import PCA
pca = PCA() 
pca_trans = pca.fit_transform(Datos)
varianza_expl = pca.explained_variance_ratio_
varianza = pca.explained_variance_
print("Valores propios: ", varianza)
```

```python
print("Vectores propios: \n", pca.components_)
```

Tomando el explained variance ratio se puede evidenciar que porcentaje del dataset que es posible explicar el porcentaje de datos del dataset que pueden ser explicados con cada dimensión

```python
print("Ratio de los valores propios:" ,varianza_expl)
```

```python
cumulative_explained_variance_plot(varianza_expl)
```

De la grafica se observa que con 1 componente es posible explicar el 98.6% de los datos, es decir que los otros 2 componentes practicamente no influyen en el dataset. En caso de querer usar un modelo de clasificación o regresión, con una dimensión sería suficiente.

# Prueba en modelos

Carga de imagenes de ejemplo

```python
import cv2
cancer = cv2.imread("fig/Cancer (1).jpg", 0)
not_cancer = cv2.imread("fig/Not Cancer  (1).jpeg", 0)
```

Carga del conjunto de datos de las imágenes

```python
import pandas as pd
df = pd.read_csv("images.csv")
df.head()
```

Binarización de las variables categóricas

```python
# Codificación de las labels
from sklearn import preprocessing
lb = preprocessing.LabelBinarizer()
df["type"] = lb.fit_transform(df["type"])
df.head()
```

División en X, y

```python
sc_x = StandardScaler()
X = sc_x.fit_transform(df.drop(["type"], axis=1))
y = df["type"]
```

## PCA

Implementación de PCA

```python
pca = PCA() 
pca.fit(X)
varianza_expl = pca.explained_variance_ratio_
varianza = pca.explained_variance_
```

```python
cumulative_explained_variance_plot(varianza_expl)
```

```python
suma = 0
for i in range(len(varianza_expl)):
    suma = suma + varianza_expl[i]
    if suma >= 0.99:
        print(i)
        break
n_components = i
n_components
```

De la grafica se puede observar que lo ideal es usar 2707 componentes

```python
pca = PCA(n_components = n_components) 
X_trans = pca.fit_transform(X)
```

Imagenes reales vs imagenes reconstruidas a partir de PCA

```python
fig, ax = plt.subplots(2, 2,figsize=(15,15))
ax[0, 0].imshow(cancer, cmap="gray")
ax[0, 0].set_title("Cancer real")
ax[0, 0].axis("off")
ax[0, 1].imshow(pca.inverse_transform(X_trans[2087]).reshape(128,128), cmap="gray")
ax[0, 1].set_title("Cancer reconstruido")
ax[0, 1].axis("off")

ax[1, 0].imshow(not_cancer, cmap="gray")
ax[1, 0].set_title("Sano real")
ax[1, 0].axis("off")
ax[1, 1].imshow(pca.inverse_transform(X_trans[0]).reshape(128,128), cmap="gray")
ax[1, 1].set_title("Sano reconstruido")
ax[1, 1].axis("off")
```

```python
# Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
XPCA_train, XPCA_test, yPCA_train, yPCA_test = train_test_split(X_trans, y, test_size=0.3, random_state=42)
```

## Modelo sin PCA

Por practicidad se estan manejando las imagenes en forma de vectores por lo que en los modelos se eliminan las capas convoluciones del modelo original, pero se mantienen las capas encargadas de clasificación. En la red neuronal original se tiene un accuracy del 98.8%.

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.metrics import CategoricalAccuracy
tf.random.set_seed(43)
```

```python
model1 = tf.keras.models.Sequential()
# Input
model1.add(tf.keras.layers.Input(shape=(128*128)))
# Clasificador
model1.add(tf.keras.layers.Dense(256, activation='relu'))
model1.add(tf.keras.layers.Dropout(0.2))
model1.add(tf.keras.layers.BatchNormalization())
model1.add(tf.keras.layers.Dense(128, activation='relu'))
model1.add(tf.keras.layers.Dropout(0.2))
model1.add(tf.keras.layers.BatchNormalization())
model1.add(tf.keras.layers.Dense(64, activation='relu'))
model1.add(tf.keras.layers.Dropout(0.2))
model1.add(tf.keras.layers.Dense(32, activation='relu'))
model1.add(tf.keras.layers.Dropout(0.2))

model1.add(tf.keras.layers.Dense(1, activation='sigmoid'))

model1.compile(loss="binary_crossentropy", optimizer=tf.optimizers.Adam(learning_rate=1e-3), metrics=["accuracy"])
```

```python
history1 = model1.fit(X_train,
                      y_train,
                     epochs=30,
                     verbose=1)
```

```python
model1.evaluate(X_test, y_test)
```

## Modelo con PCA

```python
model2 = tf.keras.models.Sequential()
# Input
model2.add(tf.keras.layers.Input(shape=(n_components)))
# Clasificador
model2.add(tf.keras.layers.Dense(256, activation='relu'))
model2.add(tf.keras.layers.Dropout(0.2))
model2.add(tf.keras.layers.BatchNormalization())
model2.add(tf.keras.layers.Dense(128, activation='relu'))
model2.add(tf.keras.layers.Dropout(0.2))
model2.add(tf.keras.layers.BatchNormalization())
model2.add(tf.keras.layers.Dense(64, activation='relu'))
model2.add(tf.keras.layers.Dropout(0.2))
model2.add(tf.keras.layers.Dense(32, activation='relu'))
model2.add(tf.keras.layers.Dropout(0.2))

model2.add(tf.keras.layers.Dense(1, activation='sigmoid'))

model2.compile(loss="binary_crossentropy", optimizer=tf.optimizers.Adam(learning_rate=1e-3), metrics=["accuracy"])
```

```python
history2 = model2.fit(XPCA_train,
                      yPCA_train,
                     epochs=30,
                     verbose=1)
```

```python
model2.evaluate(XPCA_test, yPCA_test)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.plot(history1.history["accuracy"], label="entrenamiento")
plt.title("Accuracy sin PCA"); plt.xlabel("Época"); plt.legend()
plt.subplot(122)
plt.plot(history2.history["accuracy"], label="entrenamiento")
plt.title("Accuracy con PCA"); plt.xlabel("Época"); plt.legend()
```

Cuando se comparan los modelos con y sin PCA se evidencias diferencias de 1.4% de accuracy en el set de testing, esta diferencia tiene sentido ya que se estan tomado los componentes que explican el 99% de los datos, de forma que por ese 1% es posiblemente la causa de una ligera variación en el modelo. 
Otra cosa que se observa es que a partir de las graficas es un muy rapido crecimiento en el accuracy de entrenamiento del modelo con PCA, pues a partir de la epoca 5 cuenta con más del 99% de aciertos. Por otro lado, el modelo sin PCA llega al 99% en la epoca 13. Esto deja observar que un modelo con PCA entrena más rapido por tener un vector de entrada considerablemente menor (16.384 frente a 2.707) con resultados similares.
