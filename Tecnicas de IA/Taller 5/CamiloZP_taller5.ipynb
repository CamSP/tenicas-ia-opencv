```python
# Gráfica de la varianza explicada acumulada.
def cumulative_explained_variance_plot(expl_variance):
    cum_var_exp = np.cumsum(expl_variance)
    plt.figure(dpi = 100, figsize = (8, 6))
    plt.title('Curva acumulativa de la varianza explicada VS n° de componentes principales', 
            fontdict= dict(family ='serif', size = 16))
    plt.xlabel('Número de componentes principales',
             fontdict= dict(family ='serif', size = 14))
    plt.ylabel('Varianza explicada acumulativa',
             fontdict= dict(family ='serif', size = 14))  
    nc = np.arange(1, expl_variance.shape[0] + 1)
    plt.plot(nc, cum_var_exp, '--r')
    plt.plot(nc, cum_var_exp, 'b*', ms = 5)   
    plt.show()
```

Las caracteristicas que se extraen son:
* Número de contornos con área mayor a 40 y menor a 1000. Estos intervalos son para evitar contornos de ruido o contornos muy grandes.
* Color promedio para imágenes sin preprocesar.
* Cantidad de blanco en imágenes binarizadas.
* Promedio del área de los contornos.
* Desviación estandar de las imágenes.
* Curtosis de las imágenes.

* Varianza.
* Área del cerebro.
* Momentos de Hu:
Dado que los tumores salen en diferentes tamaños, formas, posiciones, etc. Es complicado conocer un patron para estimar cual contorno corresponde a un tumopr, a partir de esto, se toman los 3 contornos más grandes que sean menores al área del cerebro y se calculan sus respectivos momentos.

```python
def data(df):
    contours = []
    mean_color = []
    white_count = []
    mean_areas = []
    std = []
    kurt = []
    skewness = []
    var = []
    brain_area = []
    Hu1_1 = []
    Hu1_2 = []
    Hu1_3 = []
    Hu1_4 = []
    Hu1_5 = []
    Hu1_6 = []
    Hu1_7 = []
    Hu2_1 = []
    Hu2_2 = []
    Hu2_3 = []
    Hu2_4 = []
    Hu2_5 = []
    Hu2_6 = []
    Hu2_7 = []
    Hu3_1 = []
    Hu3_2 = []
    Hu3_3 = []
    Hu3_4 = []
    Hu3_5 = []
    Hu3_6 = []
    Hu3_7 = []
    
    ## Imagenes procesadas
    for i in df.index:
        ## Lectura de la imagen
        img = cv2.imread("fig/"+df["label"][i]+"/"+df["Image"][i], 0)
        img_o = cv2.imread("fig/"+df["label"][i]+"_orig/"+df["Image"][i], 0)
        
        ## Contornos
        coun, hierarchy=cv2.findContours(img, cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)
        for _, c in enumerate(coun):
            new_count = []
            areas = [cv2.contourArea(c) for c in coun]
            brain_areas = []
            for a in range(len(areas)):
                if areas[a] >= 20 and areas[a] <= 1000:
                    new_count.append(coun[a])
                if areas[a] >= 20 and areas[a] <= 10000:
                    brain_areas.append(cv2.contourArea(coun[a]))
            areas = [cv2.contourArea(c) for c in new_count]
        Hu = []
        for j in (-np.array(areas)).argsort()[:3]:
            M = cv2.moments(new_count[j])
            Hu.append(cv2.HuMoments(M))
        Hu1_1.append(Hu[0][0][0])
        Hu1_2.append(Hu[0][1][0])
        Hu1_3.append(Hu[0][2][0])
        Hu1_4.append(Hu[0][3][0])
        Hu1_5.append(Hu[0][4][0])
        Hu1_6.append(Hu[0][5][0])
        Hu1_7.append(Hu[0][6][0])
        Hu2_1.append(Hu[1][0][0])
        Hu2_2.append(Hu[1][1][0])
        Hu2_3.append(Hu[1][2][0])
        Hu2_4.append(Hu[1][3][0])
        Hu2_5.append(Hu[1][4][0])
        Hu2_6.append(Hu[1][5][0])
        Hu2_7.append(Hu[1][6][0])
        Hu3_1.append(Hu[2][0][0])
        Hu3_2.append(Hu[2][1][0])
        Hu3_3.append(Hu[2][2][0])
        Hu3_4.append(Hu[2][3][0])
        Hu3_5.append(Hu[2][4][0])
        Hu3_6.append(Hu[2][5][0])
        Hu3_7.append(Hu[2][6][0])
            
            
        # Número de contornos
        contours.append(len(new_count))
        # Área promedio de los contornos internos
        mean_areas.append(np.mean(areas))
        # Área del cerebro
        brain_area.append(brain_areas[np.argmax(brain_areas)])
        # Cantidad de blanco
        white_count.append(len(img[img>0]))        
        
        
        # Promedio de color para valores > 30
        mean_color.append(img_o[img_o>30].mean())
        std.append(img_o[img_o>30].std())
        kurt.append(kurtosis(img_o[img_o>30]))
        skewness.append(skew(img_o[img_o>30]))
        var.append(img_o[img_o>30].var())
        
    ## Añadir al df
    df["Contours"] = contours
    df["mean_color"] = mean_color
    df["std"] = std
    df["mean_areas"] = mean_areas
    df["white_count"] = white_count
    df["curtosis"] = kurt
    df["skewness"] = skewness
    df["var"] = var
    df["brain_area"] = brain_area
    
    df["Hu1_1"] = Hu1_1
    df["Hu1_2"] = Hu1_2
    df["Hu1_3"] = Hu1_3
    df["Hu1_4"] = Hu1_4
    df["Hu1_5"] = Hu1_5
    df["Hu1_6"] = Hu1_6
    df["Hu1_7"] = Hu1_7
    df["Hu2_1"] = Hu2_1
    df["Hu2_2"] = Hu2_2
    df["Hu2_3"] = Hu2_3
    df["Hu2_4"] = Hu2_4
    df["Hu2_5"] = Hu2_5
    df["Hu2_6"] = Hu2_6
    df["Hu2_7"] = Hu2_7
    df["Hu3_1"] = Hu3_1
    df["Hu3_2"] = Hu3_2
    df["Hu3_3"] = Hu3_3
    df["Hu3_4"] = Hu3_4
    df["Hu3_5"] = Hu3_5
    df["Hu3_6"] = Hu3_6
    df["Hu3_7"] = Hu3_7
    
    return df
df = data(df)
df
```
